```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
  if (value=='NEW' or value=='-'):
    return np.nan
  else :
    value=str(value).split('/')
    return float(value[0])

df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate']
```

```python
df.isna().sum()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2people','listed_in(type)	':'type'},inplace=True)
```

```python
df.drop(['listed_in(city)'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['cost2people'].unique()
```

```python
def handlecomma(value):
  value=str(value)
  if ',' in value:
    value=value.replace(',','')
    return float(value)
  else:
    return float(value)
df['cost2people']=df['cost2people'].apply(handlecomma)
```

```python
  df['cost2people'].unique()
```

```python
df.head()
```

```python
rest_type=df['rest_type'].value_counts()
rest_type
```

```python
rest_typelessthan1000=rest_type[rest_type<1000]
rest_typelessthan1000
```

```python
def handle_rest_types(value):
  if(value in rest_typelessthan1000):
      return 'others'
  else:
      return value
df['rest_type']=df['rest_type'].apply(handle_rest_types)
df['rest_type'].value_counts()
```

```python
location=df['location'].value_counts()
location.unique()
```

```python
locationlessthan300=location[location<300]
```

```python
def handle_less_location(value):
  if(value in locationlessthan300):
      return 'others'
  else :
      return value
df['loaction']=df['location'].apply(handle_less_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
